# 02 — Modelagem e Avaliação

Relatório final de modelagem: baseline vs. LightGBM, calibração, avaliação por
subgrupo (fairness) e interpretabilidade (SHAP). Reutiliza os módulos de `src/`.

```python
import numpy as np

from data.loader import load_processed
from data.splitter import split_data
from evaluation.evaluator import cross_validate_model, evaluate_holdout
from models.factory import build_model
from visualization.theme import apply_theme

RANDOM_SEED = 42
rng = np.random.default_rng(RANDOM_SEED)
rng.normal()
apply_theme()
```

```python
split = split_data(load_processed())

baseline = build_model("logistic")
print(cross_validate_model(baseline, split.x_train, split.y_train).summary())

model = build_model("lightgbm")
print(cross_validate_model(model, split.x_train, split.y_train).summary())
```

```python
model.fit(split.x_train, split.y_train)
evaluate_holdout(model, split.x_test, split.y_test)
```
